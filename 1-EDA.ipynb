```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv("PremData.csv")
```

```python
# Scatterplot with a regression line showing the correlation between Average Market Value and Points
sns.regplot(data=data, x="Average_Market_Value", y="Points", ci=None, scatter_kws={'alpha':0.5})
plt.title("Correlation between Average Market Value and Points with Line of Best Fit")
plt.xlabel("Average Market Value")
plt.ylabel("Points")

# Calculating and displaying the r-squared value
slope, intercept, r_value, p_value, std_err = linregress(data["Average_Market_Value"], data["Points"])
plt.text(0.05, 0.95, f'R² = {r_value**2:.2f}', ha='left', va='center', transform=plt.gca().transAxes)

plt.show()
```

Data Transformation:  
One effect that using Average Market Value may have while aiming to predict points will be that inflation will play a role over the last twenty years. To combat this, lets say the team with the highest average market value each season will have a value of 1.0. Each team in the season will have a market value that is a proportion of itself compared to the most expensive team. 

```python
relativeValue = []

#Loop through each row
for index, team in data.iterrows():
    
    #Obtain which season we are looking at
    season = team['Year']
    
    #Create a new dataframe with just this season
    teamseason = data[data['Year'] == season]
    
    #Find the max value
    maxvalue = teamseason['Average_Market_Value'].max()
    
    #Divide this row's value by the max value for the season
    tempRelativeValue = team['Average_Market_Value']/maxvalue
    
    #Append it to our list
    relativeValue.append(tempRelativeValue)

#Add list to new column in main dataframe
data["Normalized_Average_Market_Value"] = relativeValue

#view the new table
data
```

```python
# Scatterplot with a regression line showing the correlation between Average Market Value and Points
sns.regplot(data=data, x="Normalized_Average_Market_Value", y="Points", ci=None, scatter_kws={'alpha':0.5})
plt.title("Correlation between Normalized Average Market Value and Points with Line of Best Fit")
plt.xlabel("Normalized Average Market Value")
plt.ylabel("Points")

# Calculating and displaying the r-squared value
slope, intercept, r_value, p_value, std_err = linregress(data["Normalized_Average_Market_Value"], data["Points"])
plt.text(0.05, 0.95, f'R² = {r_value**2:.2f}', ha='left', va='center', transform=plt.gca().transAxes)

plt.show()
```

Observation:  
r2 value between Average Market Value and Points: 0.53  
r2 value between Normalized Average Market Value and Points: 0.84  
This shows that inflation had a large part to play in this trend.
